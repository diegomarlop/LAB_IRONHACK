```python
db3 = db3[['Customer', 'State','Gender','Education','Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
db3
```

```python
#Los headers del db3 siguen siendo son diferentes a los del nuevo db4 por lo cual hay que estandarizarlos para poder concatenar
cols = []
for i in range(len(db3.columns)):
  cols.append(db3.columns[i].lower())

db3.columns = cols
db3
```

```python
#Siguen siendo diferentes db3 y db4 por lo que hay que seguir estandarizando 
db4.columns = ['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']

db4
```

```python
# Ya que estan estandarizadas ya se pueden concatenar db4(file1+file2) con db3(file3)
#creando una nueva db5
db5 = pd.concat([db4,db3],axis=0)
db5
```

```python
# se observa que hay 12074 filas y los indices solo marca hasta 7069 por lo que hay que estandarizar
db5 = db5.reset_index(drop=True)
db5
```

```python
#borraremos los duplicados
db5 = db5.drop_duplicates()
db5
```

```python
#Seguimos estandarizando ya sin los duplicados 
db5 = db5.reset_index(drop=True)
db5
```

```python
# Vemos los tipos de datos que tiene los headers
db5.dtypes
```

```python
# Acomodamos los headers para que los int se que den hasta el final
db5 = db5[['customer', 'state', 'gender', 'education', 'customer lifetime value', 'number of open complaints',
       'policy type', 'vehicle class','monthly premium auto','income', 'total claim amount']]
db5
```

```python
# Borraremos el header de 'education' y 'number of open complaints'
db5 = db5.drop(['education','number of open complaints'],axis=1)
```

```python
db5
```

```python
#Filtramos los customers que tienen un ingreso menor o igual a 0
db5 = db5[(db5['income']<=0)]
db5
```

```python
#manera para de quitar los % en customer lifetime value
#db5['customer lifetime value']=db5['customer lifetime value'].replace('%',"")
```

```python
#Hay que quitar el % de la columna usando replace 
db5['customer lifetime value'] = db5['customer lifetime value'].replace({'%':''}, regex=True)
#Hay que cambiar el tipo de dato que tiene la columna
db5['customer lifetime value']= pd.to_numeric(db5['customer lifetime value'],errors='ignore')
db5.dtypes
```

```python
#estandarizamos los indices para que coincidan
db5 = db5.reset_index(drop=True)
db5
```

```python
#Multiplicamos los valores por 100 usando un cilco for 
for i in range(len(db5['customer lifetime value'])):
  db5['customer lifetime value'][i] = db5['customer lifetime value'	][i]*100

db5
```
