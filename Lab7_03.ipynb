```python
import pandas as pd 
import numpy as np 
import warnings 
warnings.filterwarnings('ignore')
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('breast-cancer.csv',header=None)
data.head()
```

```python
total_cols = ['id', 'outcome', 'time', 'cell_1_radius', 'cell_1_texture', 'cell_1_perimiter','cell_1_area', 'cell_1_smoothness', 'cell_1_compactness', 'cell_1_concavity', 'cell_1_concave_points', 'cell_1_symmetry', 'cell_1_fractal_dimension', 'cell_2_radius', 'cell_2_texture', 'cell_2_perimiter', 'cell_2_area', 'cell_2_smoothness', 'cell_2_compactness', 'cell_2_concavity',        'cell_2_concave_points', 'cell_2_symmetry', 'cell_2_fractal_dimension', 'cell_3_radius', 'cell_3_texture', 'cell_3_perimiter', 'cell_3_area', 'cell_3_smoothness','cell_3_compactness', 'cell_3_concavity', 'cell_3_concave_points', 'cell_3_symmetry', 'cell_3_fractal_dimension', 'tumor_size', 'lymph_status']
```

```python
data.columns = total_cols
data.head()
```

```python
numerics = data.select_dtypes(np.number)
numerics.head()
```

```python
drop_list = ['id','time']
numerics = numerics.drop(columns=drop_list)
```

```python
from sklearn.preprocessing import StandardScaler

transformer = StandardScaler().fit(numerics)
scaled = transformer.transform(numerics)
scaled = pd.DataFrame(scaled)
scaled.columns = numerics.columns 
scaled.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(scaled)
```

```python
pca.components_
```

```python
variations = pca.explained_variance_ratio_*100
variations = np.round(variations, decimals=1)
variations
```

```python
plot_labels = ['PC'+str(i) for i in range(1,len(variations)+1)]
```

```python
from matplotlib.pyplot import figure
figure(figsize=(20,5))
plt.bar(x = plot_labels, height=variations/sum(variations))
plt.xlabel('Principal Components')
plt.ylabel('Variance Explained')
plt.title('PCA Scree Plot')
```
